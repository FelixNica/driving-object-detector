```python
from detectors import dod_beta_6 as arch
import loss
import training
import batchers
from keras.optimizers import SGD, Adam
from keras.models import load_model

detector = arch.DrivingObjectDetector()
```

```python
data_flow_train = batchers.LocalBatcher(data_path='local/data/processed/train_drv_new/',
                                        batch_size=4,
                                        restart_after=None)

data_flow_test = batchers.LocalBatcher(data_path='local/data/processed/val_drv_new/',
                                       batch_size=4,
                                       restart_after=None)
```

```python
detector.model = load_model('local/exp/1/model_states/state_final',
                            custom_objects={'loss_function': loss.get_function(detector.anchors)})

detector.model.compile(optimizer=SGD(lr=0.0001, momentum=0.3),
                       loss=loss.get_function(detector.anchors))
```

```python
training.start_session(experiment_path='local/exp/1',
                       detector=detector,
                       data_flow_train=data_flow_train,
                       data_flow_test=data_flow_test,
                       epochs=1000,
                       test_only=True)
```
